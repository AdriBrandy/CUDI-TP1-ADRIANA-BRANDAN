## Selección de datos

```python
# capital federal, venta, dolares

filtro = (df['place_l2'] == 'Capital Federal') & \
         (df['operation'] == 'Venta') & \
         (df['property_currency'] == 'USD') & \
         (df['property_type'].isin(['Casa', 'PH', 'Departamento']))

df2 = df[filtro]

# Muestra los resultados
df2
```

```python
max_price_depto = df_depto["property_price"].max()
max_price_depto
```

```python
q1 = df_depto["property_price"].quantile(0.25)
q2 = df_depto["property_price"].quantile(0.5)
q3 = df_depto["property_price"].quantile(0.75)
(q1, q2, q3)
```

```python
max = q3 + (q3-q1) * 1.5
max
```

```python
precios_atipicos = df_depto[df_depto["property_price"] >= max ]
precios_atipicos
```

```python
media_precio_atipico = precios_atipicos["property_price"].mean()
```

```python
precios_atipicos[precios_atipicos["property_price"] >= media_precio_atipico ]
```

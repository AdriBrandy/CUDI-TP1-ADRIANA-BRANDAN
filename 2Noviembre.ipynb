1. **hotel**: Tipo de hotel (por ejemplo, "City Hotel" o "Resort Hotel").
2. **lead_time**: NÃºmero de dÃ­as que transcurrieron entre la fecha de la reserva y la fecha de llegada.
3. **arrival_date_year**: AÃ±o de llegada.
4. **arrival_date_month**: Mes de llegada.
5. **arrival_date_week_number**: NÃºmero de la semana del aÃ±o para la fecha de llegada.
6. **arrival_date_day_of_month**: DÃ­a del mes de llegada.
7. **stays_in_weekend_nights**: NÃºmero de noches de fin de semana (sÃ¡bado o domingo) que el huÃ©sped se ha alojado o reservado para alojarse en el hotel.
8. **stays_in_week_nights**: NÃºmero de noches entre semana (de lunes a viernes) que el huÃ©sped se ha alojado o reservado para alojarse en el hotel.
9. **adults**: NÃºmero de adultos.
10. **children**: NÃºmero de niÃ±os.
11. **babies**: NÃºmero de bebÃ©s.
12. **meal**: Tipo de comida reservada. Las categorÃ­as son Undefined/SC (sin comida), BB (cama y desayuno), HB (media pensiÃ³n), FB (pensiÃ³n completa).
13. **country**: PaÃ­s de origen del cliente.
14. **market_segment**: Segmento de mercado al que pertenece la fuente de reserva.
15. **distribution_channel**: Canal de distribuciÃ³n de la reserva.
16. **is_repeated_guest**: Valor indicando si la persona fue un huÃ©sped repetido (1) o no (0).
17. **previous_cancellations**: NÃºmero de reservas anteriores que el cliente cancelÃ³ antes de la reserva actual.
18. **previous_bookings_not_canceled**: NÃºmero de reservas anteriores que no fueron canceladas por el cliente.
19. **reserved_room_type**: CÃ³digo de tipo de habitaciÃ³n reservada.
20. **assigned_room_type**: CÃ³digo para el tipo de habitaciÃ³n asignada. Si difiere de la reservada, esto indica una actualizaciÃ³n de la habitaciÃ³n.
21. **booking_changes**: NÃºmero de cambios realizados en la reserva desde la entrada hasta la confirmaciÃ³n de la llegada o cancelaciÃ³n.
22. **deposit_type**: IndicaciÃ³n de si el cliente hizo un depÃ³sito para garantizar la reserva. Esta variable puede contener tres categorÃ­as: "No Deposit" (sin depÃ³sito), "Refundable" (reembolsable) y "Non-Refund" (no reembolsable).
23. **agent**: ID del agente de viajes que realizÃ³ la reserva.
24. **company**: ID de la empresa o entidad que realizÃ³ la reserva o responsable de pagar la reserva.
25. **days_in_waiting_list**: NÃºmero de dÃ­as que la reserva estuvo en lista de espera antes de ser confirmada al cliente.
26. **customer_type**: Tipo de reserva, asumiendo una de cuatro categorÃ­as: Contract (cuando la reserva tiene un contrato asociado), Group (cuando la reserva estÃ¡ asociada a un grupo), Transient (cuando la reserva no forma parte de un contrato o grupo), y Transient-party (cuando la reserva es Transient, pero estÃ¡ asociada a cierta reserva Transient).
27. **adr** (tarifa diaria promedio): Calculada dividiendo la suma de todas las transacciones de alojamiento por el nÃºmero total de noches de estancia.
28. **required_car_parking_spaces**: NÃºmero de plazas de aparcamiento requeridas por el cliente.
29. **total_of_special_requests**: NÃºmero de pedidos especiales realizados por el cliente (por ejemplo, cama de matrimonio o habitaciÃ³n alta).
30. **id**: Identificador Ãºnico de la reserva.
31. **is_canceled**: Valor indicando si la reserva fue cancelada (1) o no (0).
